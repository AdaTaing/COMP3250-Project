# Combine Simulation Datasets

This notebook merges all 6 simulation output files into a single combined CSV file.

Instead of importing 6 separate files into Power BI, we import one file and filter by `Strategy` and `Portfolio_Type`.


```python
import pandas as pd
import os

# ── Folder path ───────────────────────────────────────────────────────────────
SIMULATIONS_DIR = "../data/simulations"
OUTPUT_FILE = os.path.join(SIMULATIONS_DIR, "combined_simulations.csv")

SIMULATION_FILES = [
    "single_buy_hold.csv",
    "single_timing.csv",
    "equal_buy_hold.csv",
    "equal_timing.csv",
    "custom_buy_hold.csv",
    "custom_timing.csv"
]

print(f"Simulations folder: {os.path.abspath(SIMULATIONS_DIR)}")
print(f"Output file: {os.path.abspath(OUTPUT_FILE)}")
```

## 2. Load All 6 Simulation Files

Load each file and print a quick summary so we can confirm they all loaded correctly before combining.

```python
loaded_files = {}

for filename in SIMULATION_FILES:
    path = os.path.join(SIMULATIONS_DIR, filename)
    df = pd.read_csv(path, parse_dates=["Date"])

    loaded_files[filename] = df

    # Print a summary for each file
    strategy  = df["Strategy"].iloc[0]
    portfolio = df["Portfolio_Type"].iloc[0]
    print(f"{filename}")
    print(f"Rows: {len(df)}")
    print(f"Strategy: {strategy}")
    print(f"Portfolio_Type: {portfolio}")
    print(f"Columns:  {len(df.columns)}")
```

## 3. Combine CSVs into one dataset

```python
combined_sims = pd.concat(loaded_files, ignore_index=True)

print(f"Combined dataset")
print(f"  Total rows: {len(combined_sims)}")
print(f"  Columns:    {list(combined_sims.columns)}")
print(f"\nRow counts per simulation:")
print(combined_sims.groupby(["Portfolio_Type", "Strategy"]).size().to_string())
```

## 4. Save `combined_simulations.csv`

```python
combined_sims.to_csv(OUTPUT_FILE, index=False)

print(f"Saved combined_simulations.csv")
print(f"Path: {os.path.abspath(OUTPUT_FILE)}")
print(f"Rows: {len(combined_sims)}")
print(f"Columns: {len(combined_sims.columns)}")
```
